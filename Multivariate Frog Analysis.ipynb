```python
# create a new column for year

new_frog_data['year'] = new_frog_data.eventDate.dt.year
new_frog_data
```

```python
# create a new dataframe for year and the means of tmax_mean and tmin_mean in each year

year_vs_temp = new_frog_data.groupby('year')['tmax_mean', 'tmin_mean'].mean().reset_index()
year_vs_temp
```

### Year against tmax_mean

```python
# representing year against tmax_mean on a time series

plt.figure(figsize = (12, 5))

plt.plot(year_vs_temp['year'], year_vs_temp['tmax_mean'])
plt.title('A Plot Of Year Against tmax_mean', size = 16)
plt.xlabel('year', size = 12)
plt.ylabel('tmax_mean', size = 12)
plt.show()
```

The maximum temperature rose and fell sharply from 1754 to late 18th century respectively. It rose sharply again from the late 18th century and plateaued from the first half of 19th century to the late 1850s. The temperature began to fluctuate sharply from the early years in the second half of the 19th century throughout the 20th century. In the 21st century the fluctuations have relatively been steady.

### Year against tmin_mean

```python
# representing year against tmin_mean on a time series

plt.figure(figsize = (12, 5))

plt.plot(year_vs_temp['year'], year_vs_temp['tmin_mean'])
plt.title('A Plot Of Year Against tmin_mean', size = 16)
plt.xlabel('year', size = 12)
plt.ylabel('tmin_mean', size = 12)
plt.show()
```

The minimum temperature rose sharply from 1754 till the middle of the second half of the 18th century. It then dropped sharply till the early 1800s where it rose steadily into the early years of the second half of the 19th century. The temperature began to fluctuate from the latter years in the second half of the 19th century throughout the 20th century into the 21st century where it has relatively been steady.

### Years against tmax_min and tmin_mean

```python
# create a new dataframe for year and the means of tmax_mean and tmin_mean in each year

year_vs_temp = new_frog_data.groupby('year')['tmax_mean', 'tmin_mean'].mean().reset_index()
year_vs_temp
```

```python
# representing the year, tmax_mean and tmin_mean on a scatter plot.

plt.figure(figsize = (12, 6))

plt.scatter(year_vs_temp['year'], year_vs_temp['tmax_mean'], c = 'r')
plt.scatter(year_vs_temp['year'], year_vs_temp['tmin_mean'], c = 'g')
plt.title('A Scatter plot of Year against Temperature', size = 14)
plt.xlabel('year')
plt.ylabel('temperature')
plt.show()
```

It can be observed that the maximum and minimum temperatures intersected at some points, mostly around room temperature (20 to 25) from the late 19th century to the early 1950s. This means certain frogs preferred to live in neither hot nor cold habitats at that time. It can also be observed that the temperatures recorded from the early 1950s to the 21st century do not intersect at any point and have become relatively steady (20 to 25 for the maximum temperature and 9 to 14 for the minimum temperature). This may be due to the frogs changing and adapting to live in specific narrow temperature ranges as time went by. Also, a very small fraction of the frog data was recorded in the 18th century and first half of the 19th century. A chunk of the data was recorded from the second half of the 19th century through the 20th century to the 21st century.
